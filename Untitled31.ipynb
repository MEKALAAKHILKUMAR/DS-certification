
# Loading the standard libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Loading the data

```python
data_url = "https://raw.githubusercontent.com/MEKALAAKHILKUMAR/data/main/Batchwise%20Attendance%20Data%20-%20Class%201.csv"
df = pd.read_csv(data_url)
df
```

# Data proprecessing

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Type'].fillna('STUDENT', inplace=True)
df.head()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.head()
```
